### In this notebook, I have analyzed crime characteristics based on time, days and months. The types of crimes are also analyzed with respect to days for 4 years 2015, 2016, 2017 and 2018 combined

#### First we import Python libraries for dataframe and plotting

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

#### Load the Boston crime dataset first

```python
boston_crime_dataset = pd.read_csv("C:/Users/shaya/Desktop/Kaggle datasets/crimes-in-boston/crime.csv",engine='python')
boston_crime_dataset.head()
```

```python
boston_crime_dataset.shape
```

#### Following bar plot shows the total number of crimes based on their types in descending order

```python
sns.set(font_scale=2.0)
fig, axes = plt.subplots(nrows=1,ncols=1)
fig.set_size_inches(30, 30)
ax = sns.countplot(data=boston_crime_dataset,x=boston_crime_dataset['OFFENSE_CODE_GROUP'],order=boston_crime_dataset['OFFENSE_CODE_GROUP'].value_counts().index, orient="v")
axes.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

#### Here we observe which district has the highest number of crimes reported

```python
sns.set(font_scale=2.0)
fig, axes = plt.subplots(nrows=1,ncols=1)
fig.set_size_inches(30, 30)
ax = sns.countplot(data=boston_crime_dataset,x=boston_crime_dataset['DISTRICT'],order=boston_crime_dataset['DISTRICT'].value_counts().index, orient="v")
axes.set_xticklabels(ax.get_xticklabels(),rotation=90)
ax.set(xlabel='List of Districts', ylabel='Crime Count',title="Crimes by District")
```

#### Following plot shows the total number of crimes based on the hour of day

```python
sns.set(font_scale=1.0)
sns.countplot(data=boston_crime_dataset,x=boston_crime_dataset['HOUR'])
```

#### Here we load the crime stats in dataframes based on days

```python
monday_data = boston_crime_dataset[boston_crime_dataset.DAY_OF_WEEK=='Monday']
tuesday_data = boston_crime_dataset[boston_crime_dataset.DAY_OF_WEEK=='Tuesday']
wednesday_data = boston_crime_dataset[boston_crime_dataset.DAY_OF_WEEK=='Wednesday']
thursday_data = boston_crime_dataset[boston_crime_dataset.DAY_OF_WEEK=='Thursday']
friday_data = boston_crime_dataset[boston_crime_dataset.DAY_OF_WEEK=='Friday']
saturday_data = boston_crime_dataset[boston_crime_dataset.DAY_OF_WEEK=='Saturday']
sunday_data = boston_crime_dataset[boston_crime_dataset.DAY_OF_WEEK=='Sunday']
```
